```python
def visualize(string1):
    col=string1
    ct= train_ds[col].value_counts(dropna=False).sort_index()
    grouped_data = train_ds.groupby([string1, "Exited"]).size().unstack(fill_value=0)
    proportions = grouped_data.div(grouped_data.sum(axis=1), axis=0)
    fig, ax = plt.subplots(figsize=(15, 7))
    proportions.plot(kind='bar', stacked=True, ax=ax)
    plt.xlabel(string1)
    plt.ylabel('Proportion')
    plt.legend(title="exited", bbox_to_anchor=(1, 1))
    plt.show()

for i in train_col:
    if(len(train_ds[i].unique())<100):
        visualize(i)
```

## **crating A pipeline**

```python
from sklearn.pipeline import make_pipeline, Pipeline
from sklearn.feature_selection import SelectKBest, chi2
```

```python
X_train.head()
```

```python
# Define the encoding function
def encode_age(value):
    if value <= 36.44 or value > 70:
        return 0
    elif value> 36.44   & value <= 45:
        return 1
    elif value>45  & value <= 60:
        return 2
    elif value> 60 & value <= 70:
        return 2

# Create a function transformer
age_encoder = FunctionTransformer(encode_age)

# Define ColumnTransformer
trf1 = ColumnTransformer(
    transformers=[
        ('encode_age', age_encoder, ['Age']),  # Apply age encoding to the 'Age' column
        # You can add more transformers for other columns if needed
    ],
    remainder='passthrough'  # Pass through columns not specified in transformers
)
```

```python
#sex_OHE
trf2=ColumnTransformer([
    ('Emberked_encoder',OneHotEncoder(sparse_output=False, handle_unknown='ignore'),[5])
],remainder='passthrough')
```

```python
trf3=ColumnTransformer([
    ('scaler',MinMaxScaler(),slice(0,14))
])
```

```python
pipe=make_pipeline(trf1,trf2,trf3)
pipe.fit(X_train,y_train)
```
